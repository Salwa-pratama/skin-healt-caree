```python
import numpy as np
import os
# Load Gambar
from PIL import Image
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
# Load image
img_path = "images/blackhead.jpg"
# Cek path
print("Cek file gambar : ", os.path.exists(img_path))
# img = image.load_img(img_path, target_size=(128, 128))
img = Image.open(img_path)
# resize image
img_array = np.array(img)
print(img_array)
```

```python
print("Format:", img.format)
print("Mode:", img.mode)
print("Ukuran asli:", img.size)

# Tampilkan gambar asli dulu
plt.imshow(img)
plt.title("Gambar Asli")
plt.axis('off')
plt.show()
```

```python
# Info Pixel
print("Shape gambar : ", img_array.shape)
print("Nilai pixel MIN : ", img_array.min())
print("Nilai pixel MAX : ", img_array.max())
print("\n Contoh pixel pojok kiri atas (5x5):")
print(img_array[:5, :5, 0].astype(int))
```

```python
# ----------------------------
# VISUALISASI
# ----------------------------
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Gambar asli
axes[0].imshow(img_array.astype(int))
axes[0].set_title('Gambar Asli')
axes[0].axis('off')

# Channel R (Merah)
axes[1].imshow(img_array[:,:,0], cmap='Reds')
axes[1].set_title('Channel R (Merah)')
axes[1].axis('off')

# Channel G (Hijau)
axes[2].imshow(img_array[:,:,1], cmap='Greens')
axes[2].set_title('Channel G (Hijau)')
axes[2].axis('off')
# Channel B (Biru)
axes[3].imshow(img_array[:,:,2], cmap='Blues')
axes[3].set_title('Channel B (Biru)')
axes[3].axis('off')
```

```python
plt.suptitle('Cara Sistem Membaca Gambar Sebagai Pixel', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# ----------------------------
# LIHAT NILAI PIXEL MENTAH
# ----------------------------
img_normalized = img_array / 255.0
print("\n✅ Setelah normalisasi (0-1):")
print(img_normalized[:3, :3])  # 3x3 pixel pojok kiri atas
```
